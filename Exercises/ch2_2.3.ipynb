Make a square signal at 1100 Hz and make a wave that samples it at 10000 frames per second. If you plot the spectrum, you can see that most of the harmonics are aliased. When you listen to the wave, can you hear the aliased harmonics?

```python
import thinkdsp
import thinkplot
```

<h4>1100 Hz Square Signal Segement With 10000 Frames Per Second</h4>

```python
square = thinkdsp.SquareSignal(1100)
square.plot()
square_wave = square.make_wave(duration=0.5, framerate=10000)
decorate(xlabel='Time(s)')
```

```python
square_wave.make_spectrum().plot(high=5000)
decorate(xlabel='Frequency (Hz)')
```

위 스펙트럼에서 fundamental frequency는 1100 Hz 에 있는 것을 알 수 있다. Harmonic frequency 로 먼저 3300 Hz가 있고 두번째로 5500 Hz 가 있어야 하지만 4500 Hz로, 세번째 harmonic 이 되어야 할 7700 Hz은 2300Hz로 , 네번째 9900 Hz는 100Hz로 aliasing 되었다. 이 처럼 뒤에 오는 harmonic frequency 들은 모두 aliasing 된 harmonics에 overlap 되며 aliasing 된다. 

<h3>Aliased 9900 Hz To 100 Hz</h3>

```python
square_wave.make_audio()
```

```python
square_spectrum=square_wave.make_spectrum()
square_spectrum.low_pass(100)
square_spectrum.make_wave().make_audio()
```

필터링을 통해 비교해 본 결과 Aliasing 된 100 Hz 가 fundamental pitch 로 들리는 것을 알 수 있었다.
